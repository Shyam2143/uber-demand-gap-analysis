# **Project Summary -**

**Project Overview**

This project provides a comprehensive exploratory data analysis of the Uber request dataset to identify the root causes behind the significant demand-supply gap. The primary objective was to move beyond surface-level metrics and uncover specific, actionable insights that can directly inform business strategy. The methodology involved data cleaning, extensive feature engineering to create variables like time slots and trip durations, and the generation of over 20 visualizations to dissect the relationships between trip status, pickup location, and time of day.

**Key Findings: A Tale of Two Scenarios**

The analysis revealed that the core issue is not a single, general problem but a "tale of two scenarios" occurring at different times and locations.

- The first major issue is a high volume of driver cancellations for trips originating from the City during the morning rush hours (5 AM to 10 AM).

- The second is a severe shortage of cars for requests made at the Airport during the evening peak (5 PM to 10 PM).

These two problems account for the vast majority of service failures.

**Severity and Impact**

The severity of this issue is critical; over 58% of all ride requests on the platform fail, either due to cancellations or a lack of available cars. This problem is most acute when the service is needed most. Analysis of the failure rate—the percentage of unsuccessful trips per hour—showed that it soars to over 60% in the morning and approaches a staggering 80% during evening peak hours at the airport. This indicates that the service becomes fundamentally unreliable during its busiest periods, a direct driver of negative customer experiences and potential churn.

**Predictable Patterns**

Crucially, these patterns are not random; they are highly predictable and consistent across all weekdays. The demand flow clearly shows a commuter pattern: a sharp peak in requests from the City in the morning and an equally sharp peak from the Airport in the evening. This predictability is the key to solving the problem efficiently.

**Targeted Recommendations**

Based on these specific, data-backed findings, a set of targeted recommendations were developed to address the core issues directly:

- **Implement Hyper-Targeted Incentives:** Instead of generic surge pricing, the analysis supports creating location-and-time-specific bonuses. This includes a "City-to-Airport" bonus for drivers during the morning to combat cancellations and an "Airport Pickup" surcharge during evening hours that gets passed directly to drivers to increase supply where it is most needed.

- **Leverage Technology for Supply Optimization:** The platform's algorithm should be enhanced to promote "smart trip stacking," immediately offering an airport pickup to a driver who has just completed a drop-off there. Furthermore, predictive alerts can be sent to drivers in the late afternoon, showing them the high earning potential at the airport to encourage them to reposition themselves ahead of the evening rush.

- **Strategic Communication:** Manage user expectations by notifying them of potentially longer waits during known peak problem times, and use in-app data visualizations to educate drivers on the predictable demand patterns, empowering them to maximize their earnings.

**Conclusion**

This deep-dive analysis demonstrates that by understanding the specific context behind service failures, Uber can implement precise, cost-effective strategies. Moving away from a one-size-fits-all approach to a targeted, data-driven model will directly address the primary causes of customer dissatisfaction, improve service reliability, and foster sustainable business growth.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# 1. CLEANING: Convert timestamp columns to a proper datetime format.
# Using errors='coerce' will handle any entries that cannot be converted.
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'],format='mixed')
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'],format='mixed')

# 2. FEATURE ENGINEERING: Create new columns to enable deeper analysis.

# a. Extract hour and day for time-series analysis
df['request_hour'] = df['Request timestamp'].dt.hour
df['request_day_of_week'] = df['Request timestamp'].dt.day_name()

# b. Categorize hours into broader time slots for easier aggregation
def get_time_slot(hour):
    if 5 <= hour < 10:
        return 'Morning'
    elif 10 <= hour < 17:
        return 'Daytime'
    elif 17 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'
df['time_slot'] = df['request_hour'].apply(get_time_slot)

# c. Calculate trip duration in minutes for completed trips
df['trip_duration_mins'] = (df['Drop timestamp'] - df['Request timestamp']).dt.total_seconds() / 60

# Display the first 5 rows of the now-prepared DataFrame to verify the changes
print("--- Processed and Analysis-Ready DataFrame ---")
display(df.head(3))

print("\nSetup complete. The dataset is now ready for detailed analysis.")
````

### What all manipulations have you done and insights you found?

**Data Manipulations and Preparation**

To make the dataset ready for analysis, the following manipulations were performed:

- **Data Type Conversion:**  
  The **Request timestamp** and **Drop timestamp** columns were converted from text format to **datetime format**.  
  This was essential for performing any **time-series analysis**.

- **Feature Engineering:**  
  Several new, insightful columns were created to enable deeper analysis:

  - **request_hour:** Extracted the hour of the day (0–23) from the request timestamp to analyze hourly patterns.
  - **request_day_of_week:** Derived the specific day of the week (e.g., 'Monday') to check for weekly trends.
  - **time_slot:** Grouped request hours into four distinct time slots:  
    - 'Morning' (5–10 AM)  
    - 'Daytime' (10 AM–5 PM)  
    - 'Evening' (5–10 PM)  
    - 'Night'
  - **trip_duration_mins:** Calculated the trip duration in minutes for all **completed trips** to analyze trip lengths.

Analysis of the prepared dataset revealed several critical insights into the **demand-supply gap**:
